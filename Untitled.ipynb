### Data Preprocessing

```python
missing_percentage = df.isnull().sum()*100/len(df)
missing_percentage
```

#### Selecting Important features

```python
df=df[['iyear','country_txt','region_txt','city','latitude','longitude','success','suicide','attacktype1_txt','targtype1_txt',
'targsubtype1_txt','natlty1_txt','gname','weaptype1_txt','nkill','nwound']]
```

```python
df.shape
```

```python
df.head()
```

```python
missing_percentage = df.isnull().sum()*100/len(df)
missing_percentage
```

```python
df.dropna(inplace=True)
```

#### Renaming Columns

```python
df.rename(columns={'iyear':'Year','country_txt':'Attacked Country','region_txt':'Attacked Region',
                  'city':'City','suicide':'Suicide Attack','attacktype1_txt':'AttackType','targtype1_txt':'Target',
                  'targsubtype1_txt':'Target Category','natlty1_txt':'Nationality of Target','gname':'Attacking Group',
                  'weaptype1_txt':'Weapon Used','nkill':'Deaths','nwound':'Wounded'},inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df['Casualities']= df['Deaths']+ df['Wounded']
```

```python
df.head()
```

```python
df.drop(['Deaths','Wounded'],axis=1,inplace=True)
```

```python
df.head()
```

## Exploratory Data Analysis

### Q. Which year recorded maximum attacks?

```python
plt.figure(figsize=(15,10))
sns.countplot('Year',data=df)
plt.ylabel('Number of attacks',size=15)
plt.xticks(rotation=60)
plt.xlabel('Attack Year',size=15)
plt.title('Increase in Terror',size=25, fontweight="bold")
```

<b> Remarks :- </b>
<i>Eventhough measures and actions taken by the government terrorist activities are increasing year by year. In <b>2014</b> there where maximum attacks</i>


### Q.  How many terrorist attacks were successful??

```python
plt.figure(figsize=(6,4))
sns.countplot('success',data=df)
```

```python
df['success'].value_counts()
```

<b> Remarks :- </b>
<i>  Around <b>135442</b> were terrorist attacks were successful</i>

### Q. In which year there were most no. of successful attacks?

```python
plt.figure(figsize=(20,15))
sns.countplot( 'Year',hue='success', data=df)
plt.ylabel('Number of attacks',size=25)
plt.xticks(rotation=60)
plt.xlabel('Attack Year',size=25)
plt.title('Increase in Terror',size=25, fontweight="bold")
```

<b> Remarks :- </b>
 <i>  We can see in <b>2014</b> most no of successful attacks</i>

### Q.  How many of them were Suicide attacks?

### Q.  Who are the targets of these attacks?

```python
plt.figure(figsize=(20,15))
sns.countplot( 'Target', data=df)
plt.ylabel('Number of attacks',size=25)
plt.xticks(rotation=90)
plt.xlabel('Target/victims',size=25)
```

<b> Remarks :- </b>
<i>Most of these attacks were targeted on <b>Private Citizens and Property</b></i>

### Q.  Which group caused most of the attacks?

```python
plt.figure(figsize=(10,8))
sns.barplot(y=df['Attacking Group'].value_counts()[1:20].index,x=df['Attacking Group'].value_counts()[1:20].values,orient='h')
plt.xlabel('Number of attacks',size=25)
plt.xticks(rotation=90)
plt.ylabel('Attacking Group',size=25)
```

<b> Remarks :- </b>
<i>Group causing maximum attacks are <b>Taliban followed by ISIL</b></i>

### Hot Region of Attack

```python
plt.figure(figsize=(15,10))
sns.countplot( 'Attacked Region', data=df)
plt.ylabel('Number of attacks',size=25)
plt.xticks(rotation=90)

plt.xlabel('Attack Region',size=25)
plt.title('Region wise Attack',size=25, fontweight="bold")
```

<b> Remarks :- </b>
<i>We can see that most no. of attacks occured in the <b>Middle East and North Africa followed by South Asia</b></i>

### Top affected Countries

```python
plt.figure(figsize=(20,15))
sns.barplot(y=df['Attacked Country'].value_counts()[:20].index,x=df['Attacked Country'].value_counts()[:20].values,orient='h')
plt.xlabel('Number of attacks',size=25)
plt.xticks(rotation=90)
plt.ylabel('Attacked Country',size=25)
plt.title('Country wise Attack',size=25, fontweight="bold")
```

<b> Remarks :- </b>
<i><b>Iraq</b> is attacked the most no. of times followed by <b>Pakistan and Afganistan</b></i>

### Q.  Which city was attacked the most no. of times?

```python
plt.figure(figsize=(20,15))
sns.barplot(df['City'].value_counts()[:20].index,y=df['City'].value_counts()[:20].values)
plt.ylabel('Number of attacks',size=25)
plt.xticks(rotation=90)
plt.xlabel('Attacked City',size=25)
plt.title('City wise Attack',size=25, fontweight="bold")
```

### Q.  What are the most common Attacking methods used?

```python
plt.figure(figsize=(8,6))
sns.barplot(x=df['AttackType'].value_counts()[:20].index,y=df['AttackType'].value_counts()[:20].values)
plt.ylabel('Total Attack', fontsize=20)
plt.xticks(rotation=90)
plt.xlabel('Attack Type', fontsize=20)
plt.title('Attack types',size=25, fontweight="bold")
```

<b> Remarks :- </b>
<i>From the above diagram we can observe that <b>Bombing and Explosion</b> is used most no. of times.</i>
